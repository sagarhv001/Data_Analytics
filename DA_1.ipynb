We will replace the category id with the corresponding category title, as provided to us from in a json file from the original Kaggle dataset.

```python
import json
with open("/content/drive/MyDrive/IN_category_id.json") as f:
    categories = json.load(f)["items"]
cat_dict = {}
for cat in categories:
    cat_dict[int(cat["id"])] = cat["snippet"]["title"]
dataset['category_name'] = dataset['categoryId'].map(cat_dict)
```

Let's see what is the highest views for  a video each category and which channels they belong, also visualiz the distrubution

```python
view_count = dataset.groupby('category_name')['view_count'].max()
channel_view_count = dataset.groupby('category_name').agg({'view_count': 'max', 'channelTitle': 'first'})
```

```python
view_count
```

```python
channel_view_count
```

```python
channel_views = channel_view_count.groupby('channelTitle')['view_count'].sum()
channel_views = channel_views.sort_values(ascending=False)
N = 20
top_channels = channel_views[:N]
plt.figure(figsize=(12, 6))
top_channels.plot(kind='barh', color='lime')
plt.title('maximum view count and the channel title associated with each category in the dataset')
plt.xlabel('Total View Count')
plt.ylabel('Channel Title')
plt.gca().invert_yaxis()
plt.show()
```

Let's see which is the most viewed channel

```python
most_viewed_channel=dataset.groupby('channelTitle')['view_count'].sum().idxmax()
most_viewed_channel_view_count=dataset.groupby('channelTitle')['view_count'].sum().max()


print(f'The most viewed chanel is {most_viewed_channel}with {most_viewed_channel_view_count} total views')
```

Now it's time to see if any of the numeric and boolean attributes are correlated

```python
correlation_table=['categoryId', 'view_count', 'likes', 'dislikes', 'comment_count']
correlation_table=dataset[correlation_table].corr()
correlation_table
```

So,following observations are made
<li>like_count and view_count are strongly correlated</li>
<li>other data are not strongly correlated</li>
, let's put it in a more visually appealing way

```python
plt.figure(figsize=(12,12))
sns.heatmap(correlation_table, annot=True);
```
